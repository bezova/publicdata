```python
# Let us search for all DEAD events
Dead_events = search_for_events(metadata={'WORKORDER_CONDITIONBEFORECODE':'DEAD'}).to_pandas()
Dead_events.columns
```

```python
DE_short = Dead_events[['WORKORDER_NUMBER','WORKORDER_COMPLETIONDATE', 'WORKORDER_DESC', 'assetIds']].copy()

DE_short = DE_short.sort_values(by='WORKORDER_COMPLETIONDATE')

DE_short.sort_values(by='WORKORDER_COMPLETIONDATE')
DE_short = DE_short.rename(columns={'WORKORDER_COMPLETIONDATE':'date', 
                                    'WORKORDER_NUMBER':'number',
                                    'WORKORDER_DESC':'description' })

DE_short
```

```python
DE_short = DE_short.drop_duplicates(subset='number').reset_index(drop=True)
DE_short
```

```python
# Let us augment that list with the Asset Descriptions
for i in range(0, len(DE_short)):
    event = DE_short.iloc[i]
    desc = []
    for j in event.assetIds:
        asset = get_asset(j).to_pandas()
        desc.append(asset.loc['name'][0])
    DE_short.loc[i, 'assets'] = desc
```
